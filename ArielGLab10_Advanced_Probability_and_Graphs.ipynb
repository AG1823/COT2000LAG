### Exercise 1: Rolling a Die

Calculate the probability of rolling a number greater than 4 on a standard six-sided die.
Provide your solution using the theory of probability and compare with the simulation.

### Solution:
Simulate rolling a die 10,000 times and calculate the probability of rolling a number greater than 4.

```python
import random

def simulate_rolling_die(trials=20000):        #TASK: Modify the number of trials and the situation # let's roll 20k, and find the probability of a 6
    rolls_equal_six = sum([1 for _ in range(trials) if random.randint(1, 6) == 6])
    probability = rolls_equal_six / trials
    return probability

print(simulate_rolling_die())
```

## Exercise 2: Drawing a Red Card from a Deck 
Calculate the probability of drawing a red card from a standard deck of 52 playing cards.
Provide your solution using the theory of probability and compare with the simulation.

### Solution: 
Simulate drawing a card from a standard deck of 52 cards 10,000 times to calculate the probability of drawing a red card.

```python
import random

def simulate_drawing_red_card(trials=20000):  #TASK:  Modify the number of trials and the situation # again increasing to 20k trials, getting specifically a diamond card
    diamond_draws = sum([1 for _ in range(trials) if random.choice(['diamond', 'spade', 'club', 'heart']) == 'diamond'])
    probability = diamond_draws / trials
    return probability

print(simulate_drawing_red_card())
```

## Exercise 3: Picking a Prime Number from 1 to 10
Calculate the probability of randomly selecting a prime number when picking a number from 1 to 10 (inclusive).
Provide your solution using the theory of probability and compare with the simulation.

### Solution
Simulate picking a number from 1 to 10, 10,000 times and calculate the probability of picking a prime number.

```python
import random

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def simulate_picking_prime(trials=20000):  #TASK: Modify the number of trials and the situation # again 20k, looking for composite numbers
    prime_picks = sum([1 for _ in range(trials) if not is_prime(random.randint(1, 10))])
    probability = prime_picks / trials
    return probability

print(simulate_picking_prime())
```

## Exercise 4: Even Numbers in a Lottery
In a small lottery, participants choose a number from 1 to 20. Calculate the probability of a participant choosing an even number. Provide your solution using the theory of probability and compare with the simulation.

### Solution:
Simulate a lottery where participants choose a number from 1 to 20, 10,000 times. Calculate the probability of choosing an even number.

```python
import random

def simulate_lottery_even_number(trials=20000):   # Modify the number of trials and the situation # again 20k, looking fordivisibility by 3
    even_number_picks = sum([1 for _ in range(trials) if random.randint(1, 20) % 3 == 0])
    probability = even_number_picks / trials
    return probability

print(simulate_lottery_even_number())
```

Create an undirected graph representing a small network of friends. The network has five people: Alice, Bob, Carol, Dave, and Eve.
```python
import networkx as nx
import matplotlib.pyplot as plt

def create_and_plot_undirected_graph():
    # Create an undirected graph
    G = nx.Graph()
    
    #TASK: Add a new friend node called "John"  # Done -AG
    G.add_node("Alice")
    G.add_node("Bob")
    G.add_node("Carol")
    G.add_node("Dave")
    G.add_node("Eve")
    G.add_node("John")
  
    
    #TASK: John is friend to "Carol" and "Alice"  # Add John edges
    edges = [("Alice", "Bob"), ("Alice", "Carol"), ("Bob", "Carol"), ("Bob", "Dave"), ("Carol", "Dave"), ("Carol", "Eve"), ("John", "Carol"), ("John", "Alice")]
    G.add_edges_from(edges)
    
    # Draw the graph
    plt.figure(figsize=(8, 5))
    nx.draw(G, with_labels=True, node_color='skyblue', node_size=2000, edge_color='k', linewidths=2, font_size=15)
    plt.title("Undirected Friends Network")
    plt.show()

# Call the function
create_and_plot_undirected_graph()
```

## Exercise 6: Directed Graph Creation and Visualization
Create a directed graph to represent a simple social media scenario where the nodes represent users, and an edge from user A to user B means user A follows user B. The users are: Alice, Bob, Carol, Dave, and Eve. The following relationships exist: Alice follows Bob, Bob follows Carol and Eve, Carol follows Dave and Alice, and Dave follows Eve. Visualize this network, showing the sets of vertices and edges.

```python
import networkx as nx
import matplotlib.pyplot as plt

def create_and_plot_directed_graph():
    # Create a directed graph
    G = nx.DiGraph()
    
    #TASK: remove a friend node called "Eve" # Done -AG
    G.add_node("Alice")
    G.add_node("Bob")
    G.add_node("Carol")
    G.add_node("Dave")
    
    #TASK: Eve is a friend to "Bob" and "Dave"  # remove Eve edges
    edges = [("Alice", "Bob"), ("Bob", "Carol"), ("Carol", "Dave"), ("Carol", "Alice")]
    G.add_edges_from(edges)
    
    # Draw the graph
    plt.figure(figsize=(8, 5))
    nx.draw(G, with_labels=True, node_color='lightgreen', node_size=2000, edge_color='k', linewidths=2, font_size=15, arrows=True)
    plt.title("Directed Social Media Network")
    plt.show()

# Call the function
create_and_plot_directed_graph()
```

## Exercise 7: Check if two graphs are isomorphic
Modify the examples by changing values on the adjacency_matrix and try with different graphs to verify for isomorphism.

```python
# Include these libraries
import numpy as np
from itertools import permutations
import matplotlib.pyplot as plt
import networkx as nx

# Plot a graph represented by an adjacency_matrix
def plot_graph(adjacency_matrix):
    """
    Visualizes a graph given its adjacency matrix.

    Parameters:
    adjacency_matrix (np.ndarray): The adjacency matrix of the graph to visualize.
    """
    # Converts the adjacency matrix into a NetworkX graph object for visualization.
    G = nx.from_numpy_array(adjacency_matrix)
    
    # Sets the figure size for the plot.
    plt.figure(figsize=(4, 4))
    
    # Draws the graph with specified aesthetics.
    nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=2, font_size=15, pos=nx.spring_layout(G))
    
    # Sets the title of the plot.
    plt.title('Graph Visualization')
    
    # Displays the plot.
    plt.show()

# Check if two graphs are isomorphic
def are_isomorphic(G1, G2):
    """
    Checks if two graphs represented by their adjacency matrices are isomorphic.

    Parameters:
    G1, G2 (np.ndarray): Adjacency matrices of the two graphs to compare.

    Returns:
    bool: True if the graphs are isomorphic, False otherwise.
    """
    # Checks if the matrices are square and of the same size, a prerequisite for isomorphism.
    if G1.shape != G2.shape or G1.shape[0] != G1.shape[1]:
        return False

    # The number of vertices in the graphs.
    n = G1.shape[0]
    
    # Iterates over all possible permutations of vertex labels.
    for perm in permutations(range(n)):
        # Applies the current permutation to rows and columns of G1 to simulate vertex relabeling.
        permuted_G1 = G1[np.ix_(perm, perm)]
        
        # Checks if the permuted version of G1 is identical to G2, indicating isomorphism.
        if np.array_equal(permuted_G1, G2):
            return True
    
    # Returns False if no permutation resulted in identical adjacency matrices, indicating the graphs are not isomorphic.
    return False

# Example usage
G1 = np.array([[0, 2, 0, 1],          # TASK: Modify this adjancency_matrix and check with different graphs # Modified
               [1, 0, 2, 0],
               [0, 1, 0, 2],
               [1, 2, 1, 0]])

G2 = np.array([[0, 1, 1, 0],
               [1, 1, 0, 1],
               [1, 0, 1, 1],
               [2, 1, 0, 1]])

# Checks if G1 and G2 are isomorphic and prints the result.
print(are_isomorphic(G1, G2))  # Output: True or False

# Visualizes both graphs.
plot_graph(G1)
plot_graph(G2)
```
